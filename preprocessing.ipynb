```python
len(sentences)
#i have 15 sentences
```

```python
import re
def clean_text(sentences)->list:
 corpus = [] #initialize corpus as an empty list
 for i in range(len(sentences)):
    text=re.sub('[^a-zA-Z]',' ',sentences[i])#returns a string by substituting the leftmost non-overlapping occurrences of pattern in string by the replacement repl
        # re.subn()returns a tuple containing the string obtained by replacing the leftmost non-overlapping occurrences of pattern in string by the replacement repl
    text=text.lower()#covert to lowercase
    corpus.append(text)
 return corpus  

corpus=clean_text(sentences)  
corpus
```

```python
# Stemming and stop words
#apply stemming on entire paragraph
nltk.download('stopwords')
import nltk
def stemming(sentences):
    for text in sentences:
        words=nltk.word_tokenize(text)
        for word in words:
          if word not in set(stopwords.words('english')):
                print(stemmer.stem(word))
stemming(corpus)
```
